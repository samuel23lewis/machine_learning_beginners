```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

x=np.random.uniform(0,20,10)
y=np.random.uniform(0,20,10)

#this is an important predefind function from the scipy lib used for linear regression
#the variabe before the equal sign should remain the same without deleting any of them
slope,intercept,r,p,std_err=stats.linregress(x,y) 

def my_func(x):
    return slope*x+intercept #this is a function ,this  represents the y-axis in correspondence to the x value

mymodel=list(map(my_func,x)) #each value of x runs throught the function,this results with the new array with the new array for y-axis 

plt.scatter(x,y)
plt.plot(x,mymodel) #this draws the line of linear regression
plt.show()
```
