```python
import numpy as np
import matplotlib.pyplot as plt

x=np.random.uniform(0,20,10)
y=np.random.uniform(0,20,10)

print(x)
print(y)

mymodel=np.poly1d(np.polyfit(x,y,6)) #this model helps us to find polynomial
myline=np.linspace(1,22,100) #then specifies and position the line from 1 at the start to 22 at the end

plt.scatter(x,y)
plt.plot(myline,mymodel(myline)) #draws the polynomial regression line
plt.show()
```

```python
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

print(x)
print(y)

#this model is used to find r-squared
#this is to find the relation between the x and y- axis
 
mymodel=np.poly1d(np.polyfit(x,y,3))

print(r2_score(y,mymodel(x)))

speed=mymodel(17) #
print(speed)
```
